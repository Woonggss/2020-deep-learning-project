```
import glob
from PIL import Image
```

```
# img_pathes = glob.glob("@@폴더 경로 @@/*.jpg")
img_pathes = glob.glob("/content/drive/Shared drives/선빵팀 :)/프로젝트 데이터셋(06.02 수정)/휠 카테고리/솔/*.jpg") # 차 이미지가 받아져있는 경로를 넣으면 됩니다.
```

```
img_pathes
```

```
# 충운-세웅-솔 (n) 버전
i = 365 # i값 설정은 이미지 파일 저장시 0부터 저장했으면 제일 (마지막 저장번호+1) 아니면 (마지막 저장번호)
for img_path in img_pathes:
    img = Image.open(img_path)
    img_name = img_path.split("/")[-1]
    need_name = img_name.split(")")[-1]
    img_num = img_name.split(")")[0][1:]
    new_num = int(img_num)+i # 세웅이꺼 할꺼면 맨마지막 i값 설정하고 +i 해줘야함.
    new_name = str(new_num)+"-"+need_name

    img.save('/content/drive/Shared drives/선빵팀 :)/프로젝트 데이터셋(06.02 수정)/휠 카테고리/All/' + new_name)
```
